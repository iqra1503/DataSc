The Students Performance Dataset provides insights into how students perform in academics based on various socio-economic factors such as gender, parental education, and lunch type. The dataset contains around 1000 rows and 8 coloumns. The 'gender' column refers to the gender of the student (Male/Female). The 'race/ethnicity' columns refers to the Student's ethnic background (Group A to Group E). The 'parental level of education' refers to the highest education level of a student's parent. The 'lunch' column refers to the type of lunch received. The 'test preparation course' refers to whether the student completed the test prep course or not. The final 3 columns-'math score','reading score' and 'writing score' refers to the marks obtained in the individual sectors of the students.

2️⃣ Distribution and Relationship Plots (40 Marks)


1)Histogram/Distplot: Display the distribution of a numeric variable. (10 Marks)

```python
plt.title('Marks of the subject MATH obtained by the students')
plt.ylabel('Number Of Students')
sns.histplot(student_data['math score'])
```

The histogram shows the overall range of the marks obtained by the students. From the plot, we can say that the majority of the students scored marks around 65-68. More than 100 students obtained marks from this range.

2)Pairplot: Show pairwise relationships between numerical variables. (10 Marks)

```python
sns.pairplot(student_data[['math score','reading score','writing score']])
#plt.title('Relation between exam scores')
```

A positive correlation (upward trend) between reading and writing scores would suggest students who do well in reading also perform well in writing while the linear trend between math and other subjects would indicate a strong connection in student performance across subjects.

3)Heatmap: Visualize correlations between numerical features. (10 Marks)


```python
sns.heatmap(student_data[['math score','reading score','writing score']],annot=True)
```

4)Jointplot: Analyze the relationship between two variables, including marginal distributions. (10 Marks)

```python
sns.jointplot(data=student_data,x="reading score",y="writing score",kind='kde')
```

Insights and Observations:
 The plot shows that students scoring higher in writing also scores high in reading.

3️⃣ Categorical Data Visualization (20 Marks)


1)Bar Plot: Compare categories using aggregation. (10 Marks)

```python
sns.barplot(y="parental level of education", x="math score", data=student_data)
plt.title("Math Scores by Parental Education Level")
```

Insights and Observations: 
This shows the corelation between the parental level of education and the math marks obtained by the students. This shows that the students having parents with a bachelor's degree or master's degree tend to obtain higher marks than other education levels. 

2)Box Plot/Violin Plot: Visualize the distribution of a numeric variable across categories. (10 Marks)

```python
plt.title('Variation of marks based on the lunches students get')
sns.boxplot(data=student_data,x='lunch',y='math score')
```

```python
plt.title('Variation of marks based on the lunches students get')
sns.violinplot(data=student_data,x='lunch',y='math score')
```

Insights and Observations:
This shows that students getting a standard lunch tend to obtain better marks than the students getting a free/reduced lunch, which means that the quality of lunches can also affect the performance of a student
