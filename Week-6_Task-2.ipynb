1. Dataset Selection (5 Marks)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Choose a dataset from Kaggle or other online repositories.

```python
data = pd.read_csv("erasmus.csv")
data.head(5)
```

2. Basic Plots (40 Marks)

Line Plot: Show a trend or progression of a variable over time or another index. (10 Marks)

Insight & Observations: It shows the distribution of students across different countries, highlighting patterns or trends such as the concentration of students in specific regions or variations in the number of students by country.

```python
fig=plt.figure(figsize=(10,5))
plt.title('Students Studying In Different Countries')
plt.xlabel("Student_ID")
plt.ylabel("Country")
plt.plot(data['INDEX'],data['COUNTRIES'])
plt.show()
```

Bar Chart: Compare categories or groups. (10 Marks)


Insight & Observations:It compares the number of faculties associated with each student ID, helping identify variations or patterns in faculty assignments.

```python
plt.title('Students V/S Faculties')
plt.bar(data['INDEX'],data['FACULTIES'])
plt.xlabel('Student_ID')
plt.ylabel('Faculties')
plt.show()
```

Scatter Plot: Illustrate relationships between two numeric variables. (10 Marks)

Insight & Observations : It highlights the relationship between student IDs and their exam scores, helping identify performance trends, such as variations in marks.

```python
plt.title('Relationship Between Student_ID & their Marks')
plt.scatter(data['INDEX'],data['EXAM SCORE'],marker='4')
plt.xlabel('Student_ID')
plt.ylabel('Marks')
plt.show()
```

Box Plot: Summarize the distribution and identify outliers of a numeric variable. (10 Marks)

```python
plt.boxplot(data['EXAM SCORE'])
plt.show()
```

3. Customizing Plots (20 Marks)


-Add meaningful titles, labels, and legends to all the plots. (10 Marks)

-Experiment with colors and markers to make your plots more readable and visually appealing. (10 Marks)


```python
fig=plt.figure(figsize=(10,5))
plt.title('Students Studying In Different Countries')
plt.xlabel("Student_ID")
plt.ylabel("Country")
plt.plot(data['INDEX'],data['COUNTRIES'],color='orange', linewidth=2,marker='x')
plt.show()
```

```python
plt.title('Students V/S Faculties')
plt.bar(data['INDEX'],data['FACULTIES'],color='green', edgecolor='lightgreen')
plt.xlabel('Student_ID')
plt.ylabel('Faculties')
plt.grid(axis='y', color='gray', linestyle='--', linewidth=0.5, alpha=0.5)
plt.show()
```

```python
plt.title('Relationship Between Student_ID & their Marks')
plt.scatter(data['INDEX'],data['EXAM SCORE'], color='blue', marker='.', s=100)
plt.xlabel('Student_ID')
plt.ylabel('Marks')
plt.grid(axis='y', color='gray', linestyle='--', linewidth=0.5, alpha=0.5)
plt.show()
```

```python
plt.title('Distribution of Exam Scores')
plt.boxplot(data['EXAM SCORE'], patch_artist=True,  
    boxprops=dict(facecolor='lightblue', color='darkblue'),  
    medianprops=dict(color='red', linewidth=2),  
    whiskerprops=dict(color='darkblue', linewidth=1.5), 
    capprops=dict(color='darkblue', linewidth=1.5))  
plt.show()           
```
